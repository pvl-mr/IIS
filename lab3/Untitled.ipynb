```python
def from_str_to_intarray(column_before, data):
#     column_cat = column_before
#     column_encoder = OneHotEncoder()
#     column_encoder_cat_1hot = column_encoder.fit_transform(column_cat)
#     return column_encoder_cat_1hot.toarray()

    import pandas as pd
    from sklearn.preprocessing import LabelBinarizer

    jobs_encoder = LabelBinarizer()
    jobs_encoder.fit(data[column_before])
    transformed = jobs_encoder.transform(data[column_before])
    ohe_df = pd.DataFrame(transformed)
    data = pd.concat([data, ohe_df], axis=1).drop([column_before], axis=1)
    return data
```
